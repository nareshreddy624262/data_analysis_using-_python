# how many shows are having 10 out of 10 rating

# how many showas are having 5.0 to 8.9 ratings

```python
s = 0
for i in new_data["Rating"]:
    if (i >=5.0 and i <= 8.9):
        s+=1
print(s,"shows are having rating 5,0 to 8.9 ratings ")
```
